# Web Mapping Using Folium #

```python
import folium
```

```python
# pass a geolocation: can look it up with google maps and right-click "what's here"
# The Met
map_met = folium.Map(location=[40.779469, -73.9635], tiles='Stamen Toner', zoom_start=15)
```

## try a different tile set ##
### Folium defaults to OpenStreetMap tiles, but Stamen Watercolor, Stamen Terrain, Stamen Toner, Mapbox Bright, and Mapbox Control room tiles are built-in ###

```python
# add a marker
folium.Marker([40.786309, -73.963617], popup='Central Park Reservoir').add_to(map_met)

# add a circular marker and a popup message
folium.CircleMarker([40.779469, -73.9635], radius=50,
    popup='THE MET',
    color='#3186cc',
    fill_color='#3186cc').add_to(map_met)
```

```python
map_met
```

```python
# try another tile, let's go to the himalayas
map_ter = folium.Map(location=[28.669992, 83.931870], tiles='Stamen Terrain', zoom_start=10)
map_ter.add_child(folium.LatLngPopup())
map_ter
```

```python
# watercolor tile
map_hi = folium.Map(location=[21.529130, -157.877408],
                   tiles='Stamen Watercolor',
                   zoom_start=10)
map_hi
```

```python
# other color markers and icons
map_1 = folium.Map(location=[45.372, -121.6972],
                   zoom_start=12,
                   tiles='Stamen Terrain')
folium.Marker([45.3288, -121.6625],
              popup='Mt. Hood Meadows',
              icon=folium.Icon(icon='cloud')
             ).add_to(map_1)
folium.Marker([45.3311, -121.7113],
              popup='Timberline Lodge',
              icon=folium.Icon(color='green')
             ).add_to(map_1)
folium.Marker([45.3300, -121.6823],
              popup='Some Other Location',
              icon=folium.Icon(color='red',icon='info-sign')
              ).add_to(map_1)
map_1
```

## Click-for-marker functionality will allow for on-the-fly placement of markers: ##

```python
map_4 = folium.Map(location=[46.8527, -121.7649],
                   tiles='Cartodb Positron',
                   zoom_start=13)
folium.Marker([46.8354, -121.7325], popup='Camp Muir').add_to(map_4)
map_4.add_child(folium.ClickForMarker(popup="Waypoint"))
map_4
```
