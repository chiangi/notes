## Geocoding example. Make sure geopy package is installed ##

```python
from geopy.geocoders import Nominatim # from geopy package, need internet connection. Nominatim is a service.
```

```python
nom=Nominatim(scheme="http")
n=nom.geocode("3995 23rd St, San Francisco, CA 94114", timeout=None)
n
```

```python
type(n)
```

```python
# let's use the original df and add geocode
df1 = pandas.read_csv("supermarkets.csv")
df1["Address"]=df1["Address"] + ", " + df1["City"] + ", " + df1["State"] + ", " + df1["Country"]
df1
```

## Pandas is designed in a way that allows you to apply a method or function to all the rows of the data frame without having to write a for-loop##

```python
df1["Coordinates"]=df1["Address"].apply(nom.geocode) # apply the Nominatim geocode method
df1
```

```python
df1.Coordinates[0].latitude
```

```python
df1["Latitude"]=df1["Coordinates"].apply(lambda x: x.latitude if x !=None else None) # use in-line function
#ignore the None coordinate
df1["Longitude"]=df1["Coordinates"].apply(lambda x: x.longitude if x !=None else None)
df1
```
